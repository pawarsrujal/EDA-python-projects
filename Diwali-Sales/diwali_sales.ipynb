```python
import numpy as np
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.columns
```

```python
df.describe()
```

```python
df.describe(include='all')
```

```python
p=sns.countplot(x='Gender',data=df)

for bars in p.containers:
    p.bar_label(bars)
```

```python
sales_gen= df.groupby(['Gender'],as_index=False)['Amount'].sum()
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
sns.countplot(x='Age Group',data=df,palette='Set2')
```

```python
sns.countplot(x='Age Group',data=df,hue='Gender')
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
colors = sns.color_palette("Set3", 10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=sales_state,hue='State',palette=colors)
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
colors = sns.color_palette("Set3", 10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=sales_state,hue='State',palette=colors)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
colors = sns.color_palette("Set3", 2)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status',y='Amount',data=sales_state,hue='Gender',palette=colors)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
x=sns.countplot(data=df,x='Occupation')
for bars in x.containers:
    x.bar_label(bars)
```

```python
sales_state=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
colors = sns.color_palette("Set3", 10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation',y='Amount',data=sales_state,hue='Occupation',palette=colors)
```

```python
sns.set(rc={'figure.figsize':(25,5)})
x=sns.countplot(data=df,x='Product_Category')
for bars in x.containers:
    x.bar_label(bars)
```

```python
sales_state=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
colors = sns.color_palette("Set3", 10)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x='Product_Category',y='Amount',data=sales_state,hue='Product_Category',palette=colors)
```
